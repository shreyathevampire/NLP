```
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from collections import Counter
```

```
def unique_keys(data):
  df = np.array(data)
  # print(df.shape[1])

  dictionary = []
  lists = []
  for i in range(0,df.shape[1]):
    dictval = {}
    set1= set()
    for j in range(len(df)):
      set1.add(df[j][i])
    count=0
    li = []
    if i == 6:
      set1.add('n')
      set1.add('d')
    if i == 7:
      set1.add('d')
    if i == 11:
      set1.add('u')
      set1.add('z')
    if i == 16:
      set1.add('u')
    if i == 19:
      set1.add('c')
      set1.add('s')
      set1.add('z')
    val = 0
    for j in set1:
      dictval[j] = val
      li.append((j,val))
      val += 1
      count +=1
    lists.append(dictval)
    dictionary.append(li)
  return lists
```

```
class OneHotEncoder():

  def transform(self,labels,lists):
    numeric_labels = []
    # print(labels.shape)
    for i in labels:
      # print(i)
      # print("hello")
      # print(lists[i])
      numeric_labels.append(lists[i])
    return numeric_labels

  def __init__(self):
    categories = 'auto'
  def fit_transform(self, data,lists):
    one_hot = np.empty((data.shape[0],0),dtype='int')
    nrows = data.shape[0]
    for i in range(len(lists)):
      ncols = len(lists[i])
      df = np.zeros((nrows, ncols),dtype='int')
      j = i
      tempdict = lists[i]
      for k in range(nrows):
        index = tempdict.get(data.iloc[k,j])
        df[k,index] = 1
      one_hot = np.concatenate((one_hot,df),axis=1)
      # print(one_hot.shape)


    df = np.zeros((nrows,1),dtype='int')
    tempdict = lists[i]
    for k in range(nrows):
      index = tempdict.get(data.iloc[k,j])
      df[k,0] = index
    one_hot = np.concatenate((df,one_hot),axis=1)
    # print(one_hot.shape)
    return one_hot
```

```
def knn(data, data_labels, test, k, distance_fn):
    
    final_labels = []
#     print(test.shape)
#     print(data.shape)
    for j in range(len(test)):
      distance_labels_and_indexes = []
      labels = []
      for i in range(len(data)):
  #         print(test[i,:])
          dist = distance_fn(data[i,1:], test[j,1:])
  # #         print(test[0][3])
  #         print(dist)
          distance_labels_and_indexes.append((dist, i))
      sorted_dist = sorted(distance_labels_and_indexes)
      first_k_values = sorted_dist[:k]
      # print(len(first_k_values))
      for dist,i in first_k_values:
          index = i
          labels.append(data_labels[i])
          # print("index", index, "label",data_labels[i])
      label =  mode(labels)
      print("i ", j, " label ",label)
      final_labels.append(label)
    return final_labels

def mode(labels):
    return Counter(labels).most_common(1)[0][0]

def euclidean(a, b):
  dist = np.linalg.norm(a-b)
  return dist
```
